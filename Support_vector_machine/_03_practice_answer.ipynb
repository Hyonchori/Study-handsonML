## 1. 서포트 벡터 머신의 근본 아이디어는 무엇인가요?
근본적으로 클래스 사이에 가능한 한 가장 넓은 '도로'를 내는 것. 즉 두 클래스를 구분하는 결정 경계와 샘플 사이의 마진을 가능한 가장 크게 하는 것이 목적임. 소프트 마진 분류를 할 때는 도로 안에 몇 샘플이 놓일 수 있음.  
또 하나의 아이디어는 비선형 데이터셋을 학습할 때 커널 트릭을 사용하는 것.

---
## 2. 서포트 벡터가 무엇인가요?
SVM이 결정 경계를 결정하는 데 사용하는 학습 샘플을 말함. 학습 후에는 경계를 포함해 도로에 놓인 어떤 샘플을 말함.  
직관적으로는 두 클래스의 샘플 중 가장 가까운 쌍의 샘플들을 말하는 듯. 서포트 벡터가 아닌 샘플들은 결정 경계에 영향을 주지 않음.  
예측할 때도 전체 샘플이 아닌 서포트 벡터만 사용함.

---
## 3. SVM을 사용할 때 입력값의 스케일링이 왜 중요한가요?
클래스 사이에 가능한 가장 큰 도로를 내는 것이 목적이므로 학습 데이터셋의 스케일이 맞지 않으면 크기가 작은 특성을 무시하는 경향이 생김.  
즉 도로가 제대로 만들어지지 않을 위험이 있음.

---
## 4. SVM 분류기가 샘플을 분류할 때 신뢰도 점수와 확률을 출력할 수 있나요?
테스트 샘플과 결정 경계 사이의 거리를 출력하여 이를 신뢰도 점수로 사용할 수는 있음.  
하지만 이 점수를 클래스 확률로 바로 사용할 수는 없음.  
사이킷런에서 SVM모델을 만들 때 probability=True로 설정하면 학습 후 별도의 로지스틱 회귀를 학습하여 확률을 계산할 수 있음.

---
## 5. 수백만 개의 샘플과 수백 개의 특성을 가진 학습 데이터셋에 SVM모델을 훈련시키려면 원 문제와 쌍대 문제 중 어떤 것을 사용해야 하나요?
커널 SVM은 쌍대 문제에만 사용할 수 있기 때문에 이 문제는 선형 SVM에만 해당함.  
원 문제의 계산 복잡도는 샘플수에 비례하지만, 쌍대 문제의 계산 복잡도는 샘플수의 제곱에 비례하므로 원 문제를 사용하는 것이 바람직함.

---
## 6. RBF커널을 사용해 SVM분류기를 학습했더니 학습 데이터셋에 과소적합된 것 같다. gamma를 증가시켜야 할까요, 감소시켜야 할까요? C의 경우는 어떻게 해야할까요?
과소적합됐다면 모델의 규제가 너무 큰 것일 수도 있음. 따라서 gamma와 C 모두 증가시켜 규제를 줄여야 함.
