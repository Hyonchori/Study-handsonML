# 연습문제

---
## 1. 군집을 어떻게 정의할 수 있나요? 몇 개의 군집 알고리즘을 말해보세요.
머신러닝에서 군집은 비슷한 샘플을 모으는 비지도 작업이다. 유사도 개념은 주어진 문제에 따라 다름. 예를 들어 어떤 경우에는 가까이 있는 두 샘플을 비슷하다고 생각할 수 있지만 다른 경우에는 조밀하게 모여있는 그룹에 같이 속해 있는 한 멀리 떨어져도 비슷하다고 볼 수 있음.  
k-평균, DBSCAN, 병합 군집, BIRCH, 평균-이동, 유사도 전파, 스펙트럼 군집 등이 인기있는 군집 알고리즘임.

---
## 2. 군집 알고리즘의 주요 어플리케이션은 무엇인가요?
데이터 분석, 고객 분류, 추천 시스템, 검색 엔진, 이미지 분할, 준지도 학습, 차원 축소, 이상치 탐지, 특이치 탐지 등

---
## 3. k-평균을 사용할 때 적절한 클러스터 개수를 선택할 수 있는 두 가지 기법을 설명하세요.
클러스터 개수 - 이너셔 그래프에서 엘보에 해당하는 부분을 선택.  
실루엣 점수를 그래프로 표현하여 가장 높은 클러스터 개수 선택.  
실루엣 다이어그램을 그려 모든 클러스터의 샘플수가 비슷하고 점수가 가장 높은 것을 선택.

---
## 4. 레이블 전파는 무엇인가요? 왜 이를 구현해야 하고 어떻게 구현해야 하나요?
보통 데이터셋에 사람이 직접 레이블을 부여하는 것은 많은 비용과 시간이 듬. 따라서 보통 레이블이 없는 샘플은 많고 레이블이 있는 샘플은 적음. 레이블 전파는 레이블이 있는 샘플의 일부를 레이블이 없는 비슷한 샘플에 복사하는 기법임.  
레이블을 가진 샘플의 개수를 크게 늘려 지도 학습 방법을 사용해 더 나은 성능을 낼 수 있음.(준지도 학습)  
>구현하는 한 가지 방법으로 k-평균과 같은 군집 알고리즘을 모든 샘플에 적용한 다음 군집마다 가장 많은 레이블이나 가장 대표적인 샘플(즉 센트로이드에 가장 가까운 샘플)을 찾아 동일 클러스터 안에 있는 레이블이 없는 샘플에 전파하는 것.

---
## 5. 대규모 데이터셋으로 확장할 수 있는 군집 알고리즘 두 개를 말해보세요. 밀도가 높은 지역을 찾는 군집 알고리즘 두 개는 무엇인가요?
K-평균, BIRCH는 대규모 데이터셋에 적용할 수 있음 (계산복잡도가 괜찮나 봄)  
DBSCAN, 평균-이동 알고리즘이 밀도가 높은 지역을 찾음

---
## 6. 능동 학습이 유용한 경우는 언제인가요? 어떻게 구현하나요?
레이블이 없는 샘플이 많고 레이블을 부여하는 것에 많은 비용이 들 때 유용함.  
이런 경우 무작위로 샘플을 선택해 레이블을 부여하는 것이 아니라 능동학습을 하는 것이 좋음.  
> 전문가가 학습 알고리즘과 상호작용하여 알고리즘이 요청하는 특정 샘플에 레이블을 제공하는 것.  
널리 사용되는 것은 **불확실성 샘플링** 임.

---
## 7. 이상치 탐지와 정상치 탐지의 차이는 무엇인가요?
- **이상치 탐지** : 이상치가 포함될 수 있는 데이터셋에서 알고리즘을 훈련. 이상치와 새로운 샘플 사이에 있는 이상치를 구별하는 것이 목표.  
- **특이치 탐지** : **깨끗** 하다고 가정한 데이터셋에서 알고리즘을 훈련. 새로운 샘플 사이에서 특이한 것을 감지하는 것이 목표.  
이상치 탐지에 적합한 알고리즘이 있고(아이솔레이션 포레스트), 특이치 탐지에 적합한 알고리즘도 있음(one-class SVM)

---
## 8. 가우시안 혼합이 무엇인가요? 어떤 작업에 사용할 수 있나요?
샘플들이 파라미터를 모르는 몇 개의 가우시안 분포에서 생성되었다고 가정한 확률 모델임(학습 후 샘플링도 가능)  
즉 데이터가 유한한 개수의 타원 모양 클러스터로 그룹지어 있다는 가정.(각 타원의 크기, 모양, 방향, 밀집도는 다를 수 있음)  
하지만 샘플이 어떤 클러스터에 속해 있는지는 알지 못함. 밀집도 추정, 군집, 이상치 탐지에 사용할 수 있음.

---
## 9. 가우시안 혼합 모델을 사용할 때 적절한 클러스터 개수를 찾는 두 가지 기법을 말해보세요.
클러스터 개수 - **BIC, AIC** 그래프를 그려 두 수치를 최소화하는 클러스터 개수를 선택.  
또 다른 방법은 **베이즈 가우시안 혼합 모델** 을 사용하여 클러스터 개수를 자동으로 선택.
