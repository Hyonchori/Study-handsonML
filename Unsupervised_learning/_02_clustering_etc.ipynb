# 다양한 군집 알고리즘

---
## 병합 군집
클러스터 계층을 밑바닥부터 위로 쌓아 구성.  
물 위에 작은 방울이 점점 커져서 하나의 커다란 방울이 되는 것처럼 반복마다 인접한 클러스터 쌍을 연결함.  
>**장점**  
- 대규모 샘플과 클러스터에 잘 확장됨.  
- 다양한 형태의 클러스터를 감지할 수 있음.  
- 특정 클러스터 개수를 선택하는 데 도움이 되는 유용한 클러스터 트리를 만들 수 있음.

---
## BIRCH
**balanced iterative reducing and clustering using hierarchies** 알고리즘은 특별히 대규모 데이터셋을 위해 고안됨.  
특성 개수가 너무 많지 않을 경우(20개 이하) 배치 k-평균보다 빠르고 비슷한 결과를 만듦.  
학습 과정에서 새로운 샘플을 클러스터에 빠르게 할당할 수 있는 정보를 담은 트리 구조를 만듦.  
이 트리에 모든 샘플을 저장하지 않고 제한된 메모리를 사용해 대용량 데이터셋을 다룰 수 있게 해줌.

---
## 평균-이동
먼저 각 샘플을 중심으로 하는 원을 그림. 그 다음 원마다 안에 포함된 모든 샘플의 평균을 구함. 그리고 원의 중심을 그 평균점으로 이동시킴. 모든 원이 움직이지 않을 때까지 평균-이동을 계속함.(즉 원의 중심이 포함된 샘플이 평균점일 때까지)  
평균-이동 알고리즘은 **지역의 최대 밀도를 찾을 때까지** 높은 쪽으로 원을 이동시킴. 동일한 지역에 안착한 원에 있는 모든 샘플은 동일한 클러스터가 됨. DBSCAN과 비슷한 특징이 있음.  
- 모양이나 개수에 상관없이 클러스터를 찾을 수 있음.  
- 하이퍼파라미터가 매우 적음. 밴드위스(bandwidth)라고 부르는 원 반경 딱 하나임.  
- 국부적인 밀집도 추정에 의존함.  

하지만 DBSCAN과 달리 밀집도가 불균형할 때 여러 개로 나누는 경향이 있음.  
그리고 계산 복잡도가 m제곱이라 대규모 데이터셋에 적합하지 않음.

---
## 유사도 전파
투표 방식을 사용함. 샘플은 자신을 대표할 수 있는 비슷한 샘플에 투표함. 알고리즘이 수렴하면 각 대표와 투표한 샘플이 클러스터를 형성함.  
유사도 전파는 **크기가 다른 여러 개의 클러스터를 감지할 수 있음.**  
아쉽지만 알고리즘의 계산복잡도가 m제곱이라 대규모 데이터셋에 적합하지 않음.

---
## 스펙트럼 군집
샘플 사이의 유사도 행렬을 받아 저차원 임베딩을 만듬.(즉 차원을 축소함)  
그 다음 이 저차원 공간에서 또 다른 군집 알고리즘을 사용함 (사이킷런은 k-평균을 사용함)  
스펙트럼 군집은 **복잡한 클러스터 구조를 감지하고 그래프 컷을 찾는 데 사용할 수 있음**.  
**샘플 개수가 많으면 잘 적용되지 않고, 클러스터의 크기가 매우 다르면 잘 작동하지 않음**
