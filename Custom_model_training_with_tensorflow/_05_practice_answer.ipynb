## 1. 텐서플로를 한마디로 어떻게 정의할 수 있는가? 텐서플로의 주요 특징은? 인기 있는 딥러닝 라이브러리로 어떤게 있는가?

텐서플로는 수치 계산, 특히 **대규모 머신러닝에 잘 맞도록 튜닝된 오픈 소스 라이브러리임.**  
라이브러리의 핵심 부분은 넘파이와 비슷하지만 GPU를 지원함. 또한 분산 컴퓨팅, 계산 그래프 분석과 최적화 기능을 제공.  
다른 인기있는 라이브러리로는 파이토치, MXNet, 코그니티브 툴킷, 시애노, 카페2, 체이너가 있음

## 2. 텐서플로는 넘파이를 그대로 대체할 수 있는가? 두 라이브러리의 주요 차이점은?
텐서플로가 넘파이 기능을 대부분 제공하지만 완벽한 대체제가 되지는 못함.
- 모든 함수 이름이 같지 않음 (np.sum()과 tf.reduce_sum())
- 일부 함수는 작동 방식이 완전히 같지 않음 (tf.transpose()는 텐서의 전체 복사본을 만듦. 하지만 넘파이의 T메서드는 복사가 아니라 전치된 뷰를 만듦.
- 넘파이 배열은 변경 가능하지만 텐서플로의 텐서는 만든 후 변경하지 못함. (변경 가능한 객체는 tf.Variable을 사용해야 함)

## 3. tf.range(10) 와 tf.constant(np.arange(10))의 결과는 같은가?
모두 0~9의 정수를 담은 1차원 텐서를 반환함.  
- 전자는 32비트 정수를 사용  
- 후자는 64비트 정수를 사용함.

텐서플로는 32비트가 기본이지만 넘파이는 64비트가 기본이기 때문

## 5. keras.losses.Loss 클래스를 상속하거나 일반 함수를 작성하여 사용자 정의 손실 함수를 정의할 수 있다. 언제 사용해야 하는가?
일반적으로 일반 파이썬 함수로 구현.  
하지만 일부 매개변수를 지원해야 한다면 **keras.losses.Loss** 를 상속하고 **__init__(), call()** 메서드를 구현해야 함.  
매개변수를 모델과 함께 저장하려면 **get_config()** 메서드로 구현해야 함

## 6. 비슷하게 keras.metrics.Metric 클래스를 상속하거나 함수를 정의하여 지표를 정의할 수 있다. 언제 사용해야 하는가?

일반적으로 일반 파이썬 함수로 구현  
>- 하지만 일부 매개변수를 지원해야 한다면 **keras.metrics.Metric** 클래스를 상속해야 함.

>- 또한 한 에포크 전체에 대해 이 지표를 계산하는 것과 에포크 안의 모든 배치에 대해 평균 지표를 계산하는 것이 동일하지 않으면 (예를 들어 정밀도나 재현율) **__init__(), update_state(), result()** 메서드를 구현하여 에포크동안 변화하는 지푯값을 기록해야 함.  

>- 또한 변수를 0.0으로 초기화하는 것이 필요하지 않다면 **reset_states()** 메서드도 구현해야 함.  

>-모델과 함께 저장하려면 **get_config()** 메서드도 구현해야 함.

## 11. 동적인 케라스 모델을 만들어야 할 때는 언제인가? 어떻게 만들 수 있는가? 왜 전체 모델을 동적으로 만들지 않는가?
디버깅에 유용함.  
동적인 모델을 만들려면 모델을 생성할 때 **dynamic=True** 로 지정. 또는 모델의 **compile() 메서드를 호출할 때 run_eagerly=True**로 지정.  
모델을 동적으로 만들면 케라스가 텐서플로의 그래프 장점을 활용하지 못함. 즉 학습과 추론 속도가 느려지고 계산 그래프로 내보낼 수 없어서 모델의 이식성에 제약이 생긴다고 함.
