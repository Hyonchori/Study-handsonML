```python
from _02_making_test_set import start_train_set, start_test_set
import matplotlib.pyplot as plt
```

```python
housing = start_train_set.copy()
```

훈련 세트를 손상시키지 않기 위해 복사본을 만들어 사용

```python
housing.plot(kind="scatter", x="longitude", y="latitude")
```

지리 정보(위도, 경도)를 이용하여 시각화

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1)
```

**alpha** 옵션을 이용하여 데이터가 밀집된 영역을 표현  
아마도 점이 몇 개 겹쳐야 원색이 되는지에 대한 정도인 것 같음  
  
더 두드러진 패턴을 보려면 매개변수를 다양하게 조절해야 함

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
            s=housing["population"]/100, label="population", figsize=(10, 7),
            c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
            )
plt.legend()
```

**s**는 점의 size  
**c**는 점의 색상

```python
corr_matrix = housing.corr()
corr_matrix
```

데이터셋이 크지 않으므로 모든 특성 간의 표준 상관계수 r을 구하는 **corr** 메서드를 이용  

```python
corr_matrix["median_house_value"].sort_values(ascending=False)
```

우리가 집중할 median_house_value와 다른 특성간 상관계수 출력  
해당 데이터프레임의 값을 axis에 따라 정렬해주는 **sort_values** 메서드 사용

```python
from pandas.plotting import scatter_matrix

attribs = ["median_house_value", "median_income", "total_rooms", "housing_median_age"]
scatter_matrix(housing[attribs], figsize=(13, 8))
```

각 특성 간 상관관계를 직관적으로 보기 위해 산점도로 표현해주는 **scatter_matrix** 사용  
  
자신과 자신의 상관관계는 무조건 1이기 때문에 해당 특성의 히스토그램이 표시된다.  
median_income과의 상관관계만 따로 확대해 보겠다.

```python
housing.plot(kind="scatter", x="median_income", y="median_house_value", alpha=0.1)
```

이 산점도를 보고 알 수 있는 점
> 강한 상관관계가 있음
> 앞서 확인한 가격 제한값이 수평선으로 잘 보임  
> 그런데 제한값 말고 다른 수평선도 종종 보임  
> 이런 수평선같은 데이터들은 해당 구역을 제거하는 것이 좋음
  
앞서 본 예시처럼 데이터를 탐색하면서 이상한 데이터를 확인했고, 특성 사이에서 상관관계를 발견했음.  
어떤 특성은 꼬리가 두꺼운 분포여서 데이터를 변형해야 함(예를 들어 로그 스케일로)  

```python
housing["rooms_per_household"] = housing["total_rooms"] / housing["households"]
housing["population_per_household"] = housing["population"] / housing["households"]
housing["bedrooms_per_room"] = housing["total_bedrooms"] / housing["total_rooms"]
```

여러 특성의 조합을 시도해보는 것도 좋음

```python
corr_matrix = housing.corr()
corr_matrix["median_house_value"].sort_values(ascending=False)
```

```python
import seaborn as sns

sns.heatmap(corr_matrix)
```

seaborn 패키지를 이용해 상관관계를 히트맵으로도 나타낼 수 있음

```python
sns.pairplot(housing)
plt.show()
```

```python
sns.boxplot(x="ocean_proximity", y="median_house_value", data=housing)
plt.show()
```
