## 고전적인 퍼셉트론보다 로지스틱 회귀 분류기를 일반적으로 선호하는 이유는? 퍼셉트론을 어떻게 수정하면 로지스틱 회귀 분류기와 동등하게 만들 수 있는지?
고전적인 퍼셉트론은 데이터셋이 선형적으로 구분될 때만 수렴하고 클래스 확률을 추정할 수 없음.  
반대로 로지스틱 회귀 분류기는 데이터셋이 선형적으로 구분되지 못해도 좋은 솔루션으로 수렴하고 클래스 확률을 추정할 수 있음.  
퍼셉트론의 활성 함수를 로지스틱 활성 함수로 바꾸고(또른 여러 개의 뉴런일 경우 소프트맥스로), 경사 하강법을 사용하여 학습하면 로지스틱 분류기와 동일해짐.

---
## 초창기의 다층 퍼셉트론을 학습할 때 로지스틱 활성화 함수가 왜 핵심 요소였는지?
로지스틱 활성 함수의 도함수는 어디서든 0이 아니어서 경사 하강법이 항상 경사를 따라 이동할 수 있었음. 그래서 초창기 MLP의 핵심 요소였음. 활성 함수가 계단 함수일 때는 경사가 없어서 경사 하강법을 사용 못함.

---
## 인기많은 활성 함수 세 가지는?
계단 함수, 로지스틱 함수(시그모이드), 하이퍼볼릭 탄젠트, ReLU 등

## 스팸 메일을 분류하기 위해서 출력층에 몇 개의 뉴런이 필요할지? 어떤 활성함수를 사용해야 하는지? MNIST라면 출력층의 활성 함수로 뭘 사용해야 할지?
- 하나의 뉴런, 로지스틱 활성 함수  
- 10개의 뉴런, 소프트맥스 활성 함수

## 다층 퍼셉트론에서 조정할 수 있는 하이퍼파라미터를 모두 나열하라. 과대적합되었다면 어떻게 조정해야 하는가?
- 은닉층 수, 각 은닉층의 뉴런 수, 각 은닉층과 출력층에서 사용하는 활성 함수  
> 과대적합되었다면 은닉층 수와 각 은닉층의 뉴런수를 줄여볼 수 있음.
