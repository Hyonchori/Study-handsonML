## 1. 정확히 같은 학습 데이터셋으로 다섯 개의 다른 모델을 학습해서 모두 95%의 정확도를 얻었다. 이 모델들을 연결하면 더 좋은 결과를 얻을 수 있을까요? 가능하다면 어떻게 해야 할까요?
투표 앙상블 기법을 사용하여 더 좋은 결과를 얻을 수 있음.  
모델이 서로 다를수록, 각 모델이 각각 다른 학습 샘플에서 훈련되었으면 더더욱 좋음.

## 2. 직접 투표와 간접 투표의 차이점은 무엇일까요?
직접 투표 앙상블은 각 예측기가 가장 많이 선택한 클래스를 최종 예측으로 선택하는 것  
간접 투표 앙상블은 각 예측기가 예측한 각 확률을 평균하여 가장 높은 확률의 클래스를 최종 예측으로 선택하는 것.  
일반적으로 간접 투표 방식이 더 나은 결과를 냄. 물론 각 예측기가 확률을 출력할 수 있어야 함.

## 3. 배깅 앙상블의 훈련을 여러 대의 서버에 분산시켜 속도를 높일 수 있을까요? 페이스팅 앙상블, 부스팅 앙상블, 랜덤 포레스트, 스태킹 앙상블의 경우엔 어떤가요?
배깅 앙상블의 각 예측기는 독립적이므로 여러 대의 서버에 분산하여 속도를 높일 수 있음. 페이스팅과 랜덤 포레스트도 같은 이유로 가능함.  
부스팅 앙상블은 각 예측기의 훈련이 순차적이라 분산해서 얻을 수 있는 이득이 없음

## 4. oob평가의 장점은 무엇인가요?
oob평가를 사용하면 앙상블 내의 예측기가 샘플링한 샘플을 제외한 나머지 샘플을 이용하여 검증 평가를 하므로 사전에 추가적인 검증 데이터셋을 만들지 않아도 되므로 더 많은 학습 샘플을 사용할 수 있어서 성능을 조금 더 향상시킬 수 있음.

## 5. 무엇이 엑스트라 트리를 일반 랜덤 포레스트보다 더 무작위하게 만드나요? 추가적인 무작위성이 어떻게 도움이 될까요? 엑스트라 트리는 일반 랜덤 포레스트보다 느릴까요, 빠를까요?
랜덤 포레스트는 트리가 성장할 때 각 노드에서 특성의 일부를 무작위로 선택해 분할에 사용함.  
엑스트라 트리는 이에 더해서 임계점을 찾을 때 최선의 임계점이 아닌 무작위의 임계점을 적용함.  
이 추가적인 무작위성은 규제처럼 작동함. 즉 랜덤 포레스트보다 학습 데이터에 과대적합될 위험이 적음. 그리고 훨씬 빠름.  
그러나 예측할 때는 더 빠르지도, 느리지도 않음
