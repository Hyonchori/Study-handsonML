## 1. 데이터셋의 차원을 축소하는 주 목적은 무엇인가요? 대표적인 단점은 무엇인가요?
**차원 축소의 주 목적**  
- 학습 알고리즘의 속도를 높이기 위해, 그리고 어떤 경우에는 잡음과 중복된 특성을 삭제할 수도 있어서 좋음  
- 데이터를 시각화하고 가장 중요한 특성에 대한 통찰력을 얻기 위해  
- 메모리 공간을 절약하기 위해  
  
**주요 단점**  
- 일부 정보를 손실하여 학습 알고리즘의 성능을 감소시킬 수 있음  
- 계산 비용이 높음  
- 머신러닝 파이프라인의 복잡도를 증가시킴  
- 변환된 데이터를 이해하기 어려운 경우가 많음  

---
## 2. 차원의 저주란 무엇인가요?
저차원 공간에서는 없던 많은 문제가 고차원 공간에서 일어난다는 뜻.  
머신러닝에서 무작위로 선택한 고차원 벡터는 매우 희소해서 과대적합의 위험이 큼.  
많은 양의 데이터가 있지 않으면 데이터에 있는 패턴을 잡아내기가 매우 어려워짐.

---
## 3. 데이터셋의 차원을 축소시키고 나서 이 작업을 원복할 수 있나요? 가능하다면 어떻게 가능할까요? 안된다면 왜일까요?
축소 과정에서 일부 정보는 손실됨. 즉 완벽하게 복구할 수는 없음.  
그래도 일부 알고리즘은 비교적 비슷하게 복원은 됨. 안되는 것도 있음

---
## 4. 매우 비선형적인 데이터셋의 차원을 축소할 때 PCA를 사용할 수 있을까요?
PCA는 불필요한 차원을 제거할 수 있기 떄문에 매우 비선형적이더라도 차원을 축소하는데 사용할 수 있음.  
그러나 불필요한 차원이 없다면 PCA의 차원축소는 너무 많은 정보를 잃어버리게 함.

---
## 5. 설명된 분산을 95%로 지정한 PCA를 1000개의 차원을 가진 데이터셋에 적용한다고 가정한다. 결과 데이터셋의 차원은 얼마나 될까요?
**데이터셋에 따라 다름**  
차원수에 대한 축적분산 그래프를 활용하여 정하는 것이 바람직함

---
## 6. 기본 PCA, 점진적 PCA, 랜덤 PCA, 커널 PCA는 어느 경우에 사용될까요?
- **기본 PCA** : 우선적으로 사용. 데이터셋 크기가 메모리에 맞을 떄 가능  
- **점진적 PCA** : 메모리에 담을 수 없는 대용량 데이터셋에 적합함. 하지만 기본 PCA보다는 느림. 새로운 샘플이 발생될 때마다 실시간으로 PCA를 적용해야 하는 온라인 작업에 사용 가능.  
- **랜덤 PCA** : 데이터셋이 메모리 크기에 맞고 차원을 크게 축소시킬 때 사용됨. 기본 PCA보다 훨씬 빠름.  
- **커널 PCA** : 비선형 데이터셋에 유용함.

## 7. 어떤 데이터셋에 적용한 차원 축소 알고리즘의 성능을 어떻게 평가할 수 있을까요?
직관적으로 원본 데이터셋에서 너무 많은 정보를 잃지 않고 차원을 많이 제거할 수 있다면 좋음.  
이를 측정하는 한 가지 방법은 역변환을 수행해서 재구성 오차를 측정하는 것.  
하지만 모든 차원축소 알고리즘이 역변환을 제공하지 않음. 차원 축소는 대부분 머신러닝의 전처리 단계로 사용되므로 머신러닝 알고리즘의 성능으로 평가할 수도 있음. 즉 차원 축소를 하기 전과 성능이 비슷하면 좋음.

---
## 8. 두 개의 차원 축소 알고리즘을 연결할 수 있을까요?
삽가능. PCA로 불필요한 차원을 **대폭** 제거하고 난 다음, LLE같이 훨씬 **느린** 알고리즘을 적용하는 것이 대표적인 사례라고 함.  
이런 2단계 방식은 LLE하나만 사용했을 때와 성능은 비슷하지만 속도는 훨씬 빠르다고 함.
